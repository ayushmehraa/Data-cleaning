# How many missing values do we have?
Now we know that we have missing values, Now we see how many missing values do we have.

```python
# Getting missing values per each column
missing_value_count = df.isnull().sum() # isnull() returns 1 at null and 0 at not null then sum will sum up of all 1s(null)
# looking at missing values of first 10 columns
missing_value_count[:10]
```

That seems like a lot! It might be helpful to see what percentage of the values in our dataset were missing to give us a better sense of the scale of this problem

```python
# how many cells are there
total_cell = np.product(df.shape)
total_missing = missing_value_count.sum()
# what percent of our data is missing
missing_percent = (total_missing/total_cell)*100
missing_percent
```

Wow! 24.85%, nearly quater of data is missing.In next step we will take a look at the missing values and firgure out what might be going with them.

# Figure out why data is missing
Now we will look at our data and try to figure out why it is the way it is and how that will affect our analysis.This is sometimes known as data intution.We need use our intution to figure out why value is missing? <br>
<br>
One of the most important questions we can ask ourself to help figure this out is this:<br>
            Is this value missing because it wasn't recorded or because it doesn't exist?<br>
<br>
if a value is missing becuase it doesn't exist (like the height of the oldest child of someone who doesn't have any children) then it doesn't make sense to try and guess what it might be. These values we probably do want to keep as NaN. On the other hand, if a value is missing because it wasn't recorded, then we can try to guess what it might have been based on the other values in that column and row. This is called imputation, and we'll learn how to do it next!

By looking at the documentation (https://www.kaggle.com/datasets/maxhorowitz/nflplaybyplay2009to2016), we can see that this column has information on the number of seconds left in the game when the play was made. This means that these values are probably missing because they were not recorded, rather than because they don't exist. So, it would make sense for us to try and guess what they should be rather than just leaving them as NA's. <br>

On the other hand, there are other fields, like "PenalizedTeam" that also have lot of missing fields. In this case, though, the field is missing because if there was no penalty then it doesn't make sense to say which team was penalized. For this column, it would make more sense to either leave it empty or to add a third value like "neither" and use that to replace the NA's.<br>

# Drop Missing values
<br>
If we're in a hurry or don't have a reason to figure out why our values are missing, one option you have is to just remove any rows or columns that contain missing values. (Note: <b>This is generally not recommend approch for important projects</b>! It's usually worth it to take the time to go through your data and really look at all the columns with missing values one-by-one to really get to know your dataset.<br>
<br>
If we're sure we want to drop rows with missing values, pandas does have a handy function, dropna() to help you do this

```python
# droping rows containing missig values
drop_missing_values = df.dropna()
drop_missing_values
```

Ohh it droped all rows in our dataset this is because each and every row in our dataset contain missing value.<br>
Lets try drop column containing missing value this time.

```python
#dropping columns with missing values
columns_with_dropna = df.dropna(axis=1) #axis 1 means on cloumn and 0 mean row by default it is 0.
columns_with_dropna
```

its better than empty dataset!
<br>
now lets see how much of proportion did we loose.


```python
# how much od data we lose
print(f'Columns in original dataset : {df.shape[1]}')
print(f'Columns in original dataset : {columns_with_dropna.shape[1]}')
print(f'Total Column loss {df.shape[1]-columns_with_dropna.shape[1]}')
```

Wow ! seems we did loss some quite amount of data.However we did successfully removed all NaN/null/missing values from our dataset.

# Filling in missing values automatically

Another option is to try and fill in the missing values.<br> 
For this next bit, we are getting a small sub-section of the NFL data so that it will print well.

```python
# Getting samll subset of NFL dataset df
subset_df = df.loc[:,'EPA':'Season'].head()
subset_df
```

We can use the Panda's fillna() function to fill in missing values in a dataframe for us. One option we have is to specify what we want the NaN values to be replaced with. Here, I'm saying that I would like to replace all the NaN values with 0.

```python
# replace all NaN with 0
subset_df.fillna(0)
```

we could also be a bit more savvy and replace missing values with whatever value comes directly after it in the same column. (This makes a lot of sense for datasets where the observations have some sort of logical order to them.)

```python
# replace all Na values that comes directly after it in same column.
# Then replace all remaining wiht 0.
subset_df.fillna(method = 'bfill' , axis = 0 ).fillna(0)
```
